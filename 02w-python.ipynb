```python
A = {1,2,3,4}
B = {3,4,5,6,7}
print(A.union(B)) #합집합
print(A.intersection(B)) #교집합
print(A.difference(B)) #차집합
print(A.difference(A)) #차집합
print(A.symmetric_difference(B)) # 여집합
```

```python
A = {1,2,3,4}
B = {3,4,5,6,7}
print(A | B) #합집합
print(A & B) #교집합
print(A - B) #차집합
print(B - A) #차집합
print(A ^ B) # 여집합
```
